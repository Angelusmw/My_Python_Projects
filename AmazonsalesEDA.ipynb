# EDA

```python
# Which products had the highest discount percentage
category_by_discountpercentage = (df
                 .groupby("product_name")
                 .discount_percentage.sum()
                 .reset_index()
                 .sort_values(by=['discount_percentage'], ascending=False))

print(category_by_discountpercentage)
```

```python
# Which products had the highest ratings
category_by_highestrating = (df
                 .groupby("product_name")
                 .rating.sum()
                 .reset_index()
                 .sort_values(by=['rating'], ascending=False))

print(category_by_highestrating)
```

```python
#Which Amazon products had the highest ratings count? 

#Some values had improperly positioned commas in the rating-count series 

df['rating_count'] = pd.to_numeric(df['rating_count'].str.replace(',', ''), errors='coerce')

category_by_highestproductcount = (df
                 .groupby("product_name")
                 .rating_count.sum()
                 .reset_index()
                 .sort_values(by=['rating_count'], ascending=False))
print(category_by_highestproductcount)
```

```python
#Convert rating series to numeric values

df['rating'] = pd.to_numeric(df['rating'],errors='coerce')
```

```python
#Visualization for top 10 highest rated Amazon Products

# prepare plot
plt.figure(figsize=(15,6))
sns.set(font_scale=1.5, style="whitegrid")

# set color
labels = np.array(df.product_name)
values = np.array(df.rating) 
colors = ['blue']

# set the plot
ax = sns.barplot(x="product_name", 
                 y="rating", 
                 data=df.head(10), 
                 palette = colors)

# set title and arrange x,y ticks
plt.title("Top Ten Highest-Rated Amazon Products")
for item in ax.get_xticklabels():
    item.set_rotation(90)

```
