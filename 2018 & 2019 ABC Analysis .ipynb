```python
import pandas as pd
```

```python
#reading data from the excel sheet named 'SalesItemLevel'

df = pd.read_excel(r'C:\Users\Administrator\Downloads\MILES_Data_C11BA_2023 (1).xlsb', sheet_name = 'SalesItemLevel')
df.head()
```

```python
#setting the third row's entries as default column names

header_row = df.iloc[1]
df = pd.DataFrame(df.values[2:], columns=header_row)

df.head()
```

```python
# converting the 'Total Qty','Total_Qty_2018','Total_Qty_2019' columns into float data types

df[['Total Qty','Total_Qty_2018','Total_Qty_2019']] = df[['Total Qty','Total_Qty_2018','Total_Qty_2019']].astype(float)
df.info()
```

```python
#reading data from the excel sheet named 'Products'

dfpro = pd.read_excel(r'C:\Users\Administrator\Downloads\MILES_Data_C11BA_2023 (1).xlsb', sheet_name = 'Products')
dfpro.head()
```

```python
#setting the third row's entries as default column names

header_row = dfpro.iloc[1]
dfpro = pd.DataFrame(dfpro.values[2:], columns=header_row)
```

```python
dfpro.head()
```

```python
# left joining the dataframes named 'df' and 'dfpro' on the 'Product Description' key

df_merged = pd.merge(df,dfpro,how='left',on="Product Description")
df_merged.info()
```

```python
#converting the 'Unit Selling Price' column into a float data type to allow for computations

df_merged['Unit Selling Price'] = pd.to_numeric(df_merged['Unit Selling Price'])
df_merged.info()
```

```python
#converting the 'Unit Selling Price' column into a object data type

df_merged['Unit Selling Price'] = df_merged['Unit Selling Price'].astype(object)
df_merged.info()
```

```python
#Grouping the resultant dataframe by the 'Product Description' and using the sum aggregation function

dfabc = df_merged.groupby('Product Description').sum()

dfabc.head(15)
```

```python
#Creating a new dataframe that contains the 'Product Description' and 'Unit Selling Price' columns

df_usp = dfpro[['Product Description','Unit Selling Price']]
df_usp.head()
```

```python
#inner joining dataframes 'df_usp' and 'dfabc' on the "Product Description" key

df3 = pd.merge(df_usp,dfabc,how='inner',on="Product Description")
df3.head()
```

```python
#Creating a new column in the 'df3' dataframe by multiplying 'Total_Qty_2018' and 'Unit Selling Price'

df3['2018 Annual Consumption Value'] = df3['Total_Qty_2018'] * df3['Unit Selling Price']
df3.head()
```

```python
#Creating a new column in the 'df3' dataframe by multiplying 'Total_Qty_2019' and 'Unit Selling Price'

df3['2019 Annual Consumption Value'] = df3['Total_Qty_2019'] * df3['Unit Selling Price']
df3.head()
```

```python
#Sorting the new column from the largest to the smallest values

df_2018acv = df3.sort_values(by='2018 Annual Consumption Value', ascending=False)

df_2018acv.head()
```

```python
#Sorting the new column from the largest to the smallest values

df_2019acv = df3.sort_values(by='2019 Annual Consumption Value', ascending=False)

df_2019acv.head()
```

```python
#Calculating the cumulative percentage using 2018 data and employing the cumulative percentage to place products into three categories


df_2018acv['Cumulative_PercentageACV'] = df_2018acv['2018 Annual Consumption Value'].cumsum() / df_2018acv['2018 Annual Consumption Value'].sum() *100
df_2018acv['Category'] = pd.cut(df_2018acv['Cumulative_PercentageACV'], bins=[0, 20, 50, 100.1], labels=['A', 'B', 'C'])

df_2018acv.head()
```

```python
#Converting the column named '2018 Annual Consumption Value' into float data type to allow for computations

df_2018acv['2018 Annual Consumption Value'] = df_2018acv['2018 Annual Consumption Value'].astype(float)
df_2018acv.info()
```

```python
#Calculating the total units sold in each category using the group by function. 

dfsales = df_2018acv.groupby('Category').sum()

dfsales.head(15)
```

```python
#Calculating the percentage of product sales for each of the three categories.


dfsales['Percentage_ACV'] = dfsales['2018 Annual Consumption Value'] / df_2018acv['2018 Annual Consumption Value'].sum() *100
dfsales
```

```python
#Sorting the total quantity of products sold in 2018 from the largest to the smallest

df_2018tus = df_2018acv.sort_values(by='Total_Qty_2018', ascending=False)

df_2018tus.head()
```

```python
#Calculating the cumulative percentage using 2018 data and employing the cumulative percentage to place products into three categories


df_2018tus['Cumulative_Percentage_T Units Sold'] = df_2018tus['Total_Qty_2018'].cumsum() / df_2018tus['Total_Qty_2018'].sum() *100
df_2018tus['CategoryUS'] = pd.cut(df_2018tus['Cumulative_Percentage_T Units Sold'], bins=[0, 20, 50, 100.1], labels=['A', 'B', 'C'])

df_2018tus.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
#define aesthetics for plot
color1 = 'steelblue'
color2 = 'red'
line_size = 2

#create basic bar plot
fig, ax = plt.subplots(figsize=(10, 6),dpi=200)
ax.bar(df_2018acv['Product Description'], df_2018acv['2018 Annual Consumption Value'], color=color1)
ax.set_xlabel('Product Description')
ax.set_ylabel('Annual Consumption Value', color='tab:blue')
ax.set_ylim(0, 5412035)

# Remove x-axis labels
ax.set_xticklabels([])


#add cumulative percentage line to plot
ax2 = ax.twinx()
ax2.plot(df_2018tus['Product Description'], df_2018tus['Cumulative_Percentage_T Units Sold'], color=color2, marker="D", ms=line_size)
ax2.yaxis.set_major_formatter(PercentFormatter())
ax2.set_ylabel('Cumulative Percentage (%)', color='tab:red')

#specify axis colors
ax.tick_params(axis='y', colors=color1)
ax2.tick_params(axis='y', colors=color2)

plt.title('2018 Pareto Chart')

plt.savefig('C:\\Users\\Administrator\\Downloads\\2018_pareto_chart_updated.png')

#display Pareto chart
plt.show()
```
